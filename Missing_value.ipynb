```python
#MengImport library yang dibutuhkan
import pandas as pd
import numpy as np
```

```python
#Dataset yang digunakan diambil dari Github https://github.com/Yugesh-Ravi/Exploratory-Analysis-of-Melbourne-Housing-Data.git
#dimana data tersebut kami modifikasi agar lebih mudah untuk dipahami. 
#dataset yang diambil adalah data penjualan rumah di Melbourne,data tersebut terdiri dari 201 baris dan 8 kolom 
#yang berisi tanggal, letak, alamat, tipe rumah, jumlah kamar, harga, metode pembayaran, dan penjual


df = pd.read_csv('https://raw.githubusercontent.com/rimayadwiatika11/DATASET_STATISTIK_G1A021019-DAN-G1A021021/main/Dataset_Melbourne_Housing.csv')
#kode diatas digunakan ketika kita ingin menambahkan dataset/load dataset menggunakan URL

```

```python
df.head()
#df.head()digunakan untuk menampilkan 5 data awal atau data teratas pada dataframe. 
```

```python
# melihat informasi dataset. Semua kolom memiliki 201 data non-null kecuali pada kolom price yang hanya terdapat 195 data non-null
# artinya pada kolom price terdapat data yang bernilai null/NaN
df.info()
```

```python
# Mengidentifikasi/melihat banyak data yang memiliki nilai null/NaN pada kolom "Price"
df.isnull().sum()
```

```python
#digunakan untuk mendeskripsikan nilai Q1,Q2,Q3, mean dan data lainnya pada kolom yang memiliki missing value yaitu "Price"

df['Price'].describe()
```

```python
#Mengganti Missing value menggunakan Mean (Metode Imputation)

mean = df["Price"].mean()
df["Price"] = df["Price"].replace(np.nan, mean)
df.head()

#terlihat bahwa data ke 2 dan 5 pada kolom Price yang semula NaN berubah menjadi nilai mean yaitu 1.623832e+06
```

```python
#Mengganti Missing value menggunakan Mode (Metode Imputation)

df['Price'] = df['Price'].fillna(df['Price'].mode()[0])
df.head()

#terlihat bahwa data ke 2 dan 5 pada kolom Price yang semula NaN berubah menjadi nilai mode yaitu 1801000.0
```

```python
#Mengganti Missing value menggunakan Median (Metode Imputation)

median = df["Price"].median()
df["Price"] = df["Price"].replace(np.nan, median)
df.head()

#terlihat bahwa  data ke 2 dan 5 pada kolom Price yang semula NaN berubah menjadi nilai median yaitu 1030000.0
```

```python
#menampilkan panjang data yang akan dihapus missing value-nya
print("\nPanjang Data :" , df.shape)
```

```python
#Menghapus Missing Value Pada Dataset
df.dropna(inplace=True)
print(df.isnull().sum())

#setelah dihapus Missing Value pada kolom Price menjadi 0
```

```python
#menampilkan data yang telah dihapus missing value-nya, dapat dilihat sebelumnya terdapat 201 baris data setelah dihapus menjadi 195 baris data
#artinya sebanyak 6 data yang memiliki missing value telah dihapus
print("\nPanjang Data :" , df.shape)
```
